```python
import inspect
import numpy as np
var('t')
def euler(init, fn, step=0.1, end=1):
    res = [init]
    for t in np.arange(init[0] + step, end + step, step):
        if len(inspect.getargspec(fn).args) == 2:
            res.append([round(t * 1e5) / 1e5, res[-1][1] + step *fn(res[-1][1], res[-1][0])])
        else:
            res.append([round(t * 1e5) / 1e5, res[-1][1] + step *fn(res[-1][1])])
    return res
```

```python
table(euler([0,0.2], lambda y: y*(1-y), 0.5, 3))
```
